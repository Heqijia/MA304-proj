```python
import matplotlib.pyplot as plt
import numpy as np

# 定义数据
plt.rcParams['font.family'] = 'Microsoft YaHei'

categories = ['机械行业', '化工行业', '生物医药']
values_0 = [-2.152678132678133, -1.8492162162162162, -3.366812865497076]
values_1 = [0.2960000000000001, 0.15800000000000028, -0.3820000000000011]

# 定义x轴位置
x = np.arange(len(categories))  # 类别标签的位置

# 柱状图的宽度
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, values_0, width, label='行业股票均值变化',color='green',alpha=0.8)
rects2 = ax.bar(x + width/2, values_1, width, label='投资建议top10股票均值变化',color='red',alpha=0.8)

# 添加一些文本标签
ax.set_ylabel('股票涨跌均值',fontsize=12)
ax.set_title('各行业股票涨跌均值变化',fontsize=15)
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend(fontsize=10)
ax.axhline(y=0, color='black', linestyle='--')
ax.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

plt.show()
```
